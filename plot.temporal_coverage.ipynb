```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df_GAMA = pd.read_csv('CENTRALVALLEY_NO3N_conc_0_zipcoded.csv') 
df_UCD = pd.read_csv('UCD_zipcode_filtered.csv')
print(len(df_GAMA))
df_GAMA = df_GAMA[df_GAMA['Zipcode'].isin(df_UCD['Zipcode'].unique())]
print(len(df_GAMA))
```

```python
GAMA_times = pd.to_datetime(df_GAMA['GM_SAMP_COLLECTION_DATE'], format= '%m/%d/%y')
GAMA_times.hist(bins = 40, label = 'GAMA')


df_UCD['DATE'] = pd.to_datetime(df_UCD['DATE'], format= '%m/%d/%y')
df_UCD = df_UCD[df_UCD['DATE']>= dt.datetime(1990, 1, 1)]
df_UCD = df_UCD[df_UCD['DATE']<= dt.datetime(2010, 12, 31)]
df_UCD['DATE'].hist(bins = 40, label = 'UCD')

plt.xlabel('Date')
plt.ylabel('Number of Measurements')
plt.title('Histogram of Measurements Over Time')

plt.legend(loc = 'upper left')
```
