```python
import pandas as pd
```

```python
df= pd.read_csv('CENTRALVALLEY_NO3N_conc_0_updated_zipcoded.csv')

df['GM_SAMP_COLLECTION_DATE'] = pd.to_datetime(df['GM_SAMP_COLLECTION_DATE'])

# Extract year from GM_SAMP_COLLECTION_DATE
df['Year'] = df['GM_SAMP_COLLECTION_DATE'].dt.year

all_zipcodes = df['Zipcode'].unique()

years=df['Year'].unique()
 
```
